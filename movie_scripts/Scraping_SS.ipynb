```python
import urllib3
urllib3.disable_warnings()
from bs4 import BeautifulSoup
```

```python
def format_text(soup):
    for br in soup.find_all("br"):
        br.replace_with("\n")
    return soup.text.strip().replace('\n ','\n')
```

```python
http = urllib3.PoolManager()
```

```python
tv_shows = ['30-rock', 'parks-and-recreation']
num_seasons = [7, 7]
max_eps = [22, 24]
```

```python
for tv_id in range(len(tv_shows)):
    web_pre = 'https://www.springfieldspringfield.co.uk/view_episode_scripts.php?tv-show='
    
    # Remove old text file and create a new one for that show. 
    txt_file = tv_shows[tv_id] + ".txt"
    import os
    try:
        os.remove(txt_file)
    except IOError:
        pass
    f= open(txt_file,"a+", encoding='utf-8')
    
    
    for season_num in range(1, num_seasons[tv_id] + 1):
        for ep_num in range(1, max_eps[tv_id] + 1):
            print(tv_shows[tv_id] + " Season #" + str(season_num) + " Episode #" + str(ep_num))
            webpage = web_pre + tv_shows[tv_id] + "&episode=s{:02d}e{:02d}".format(season_num, ep_num)
            response = http.request('GET', webpage)
            soup_response = BeautifulSoup(response.data)
            results = soup_response.findAll("div", {"class": "scrolling-script-container"})
            if (len(results) == 1):
                    text = format_text(results[0])
                    if (os.stat(txt_file).st_size != 0):
                        f.write("\n")
                    f.write(text)
            else:
                print("There was an error with season "  + str(season_num) + " episode " + str(ep_num))
                print("Results are " + str(results))
    f.close()
```
